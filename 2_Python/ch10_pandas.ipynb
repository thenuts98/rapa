### 4.2 loc[]을 이용한 이름(열, 행)으로 조회
- 단일행, 여러열, 행과 열을 동시에 접근
- member.loc[행(행슬라이싱), 열(열슬라이싱)]

```python
member.loc[0] # 시리즈
```

```python
#여러열
member.loc[:, 'name' : 'email'] # from : to : step
```

```python
# 특정 열만 뽑기
member.loc[:, ['name', 'email']] # 뽑을 열만 list형식으로
```

```python
# 1행부터 4행까지 name과 email
member.loc[1:4, ['name', 'email']]
```

### 4.3 iloc[]을 이용한 숫자(인덱스)로 조회
- 몇번째 행부터 몇번째 행 앞까지, 몇번째 열부터 몇번째 옆 앞까지 숫자로 접근. 음수 가능


```python
member.iloc[:, 0:3]
```

```python
# 맨 오른쪽 열을 제외하고 출력
member.iloc[:, :-1]
```

```python
#짝수번째 행의 모든 열 출력
member.iloc[::2]
```

```python
# member의 행순서를 역순 출력(맨 마지막열을 제외하고)
member.iloc[::-1, :-1]
```
